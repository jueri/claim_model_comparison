```python
import pandas as pd
import matplotlib.pyplot as plt  # type: ignore
import tikzplotlib
```

```python
data = pd.read_csv("data/wandb_export_2021-12-22T15 30 01.504+01 00.csv")
```

```python
data.dataset.unique()
```

### IBM_Debater_(R)_claim_sentences_search

```python
a = data[data.dataset=="IBM_Debater_(R)_claim_sentences_search"].pivot(columns='model', values='test_accuracy')
a = a[a.mean().sort_values().index]  # oder columns
a.rename(columns={"distilbert-base-uncased": "BERT"}, inplace=True)

plt.figure(facecolor="white", figsize=(12,8))
plt.xlabel("System name")
plt.ylabel('Result accuracy')
plt.title('Variance of results produced by model trained on the Claims and Evidence dataset, measured in accuracy')
a.boxplot()
# plt.show()
tikzplotlib.clean_figure()
tikzplotlib.save("comparision_variance.tex")
```

```python
a = data[data.dataset=="IBM_Debater_(R)_CE-ACL-2014.v0"].pivot(columns='model', values='test_accuracy')
a = a[a.mean().sort_values().index]  # oder columns

plt.figure(facecolor="white", figsize=(12,8))
plt.xlabel("Model")
plt.ylabel('accuracy')
plt.title('Dataset: IBM_Debater_(R)_CE-ACL-2014.v0 Plot')
a.boxplot()
plt.show()
```
